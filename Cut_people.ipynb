```python
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = YOLO('yolov8n.pt')
path = '/content/drive/MyDrive/trainee/dataset1.1/train/images/138.jpg'
res = model.predict(path, classes=0, single_cls=True)[0]
# Получаем рамку первого человека
frame1 = np.uint32(res.boxes.xyxy.numpy()[0])
# Получаем целое изображение
img_orig = res.orig_img
# Получаем фрагмент изображения первого человека.
# Здесь сначала идут у, потом х, цвета BGR
img = img_orig[frame1[1]:frame1[3], frame1[0]:frame1[2], ::-1]
plt.figure(figsize=(16, 16))
plt.imshow(img)
plt.show()
```
