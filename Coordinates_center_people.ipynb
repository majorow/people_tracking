```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/trainee/coordinates.csv')
np_ar = df.to_numpy(int)
x = np.uint32(np.mean((np_ar[:, 0], np_ar[:, 2]), axis=0))
y = np.uint32(np.mean((np_ar[:, 1], np_ar[:, 3]), axis=0))
hstgrm, _, _ = np.histogram2d(x, y, bins=160)
img = np.uint8((1 - hstgrm / hstgrm.max()) * 255)
img = np.expand_dims(img, axis=-1)
plt.figure(figsize=(16, 16))
plt.imshow(img, cmap='gray')
plt.show()
```

```python
hstgrm, _, _ = np.histogram2d(x, y, bins=80)
img = np.uint8((1 - hstgrm / hstgrm.max()) * 255)
img = np.expand_dims(img, axis=-1)
plt.figure(figsize=(16, 16))
plt.imshow(img, cmap='gray')
plt.show()
```
