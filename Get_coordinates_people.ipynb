```python
from ultralytics import YOLO
import numpy as np
from IPython.display import clear_output
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# COCO
model = YOLO('yolov8s.pt')
path = '/content/drive/MyDrive/trainee/dataset1.1/val/images/'
res = model.predict(path, imgsz=1280, single_cls=True, classes=0)
clear_output()
people = np.empty((0, 4))
for res1 in res: 
  result = res1.cpu().boxes.boxes.numpy()
  people = np.vstack((people, result[:, :4]))
df = pd.DataFrame(np.uint32(people), columns=['x1', 'y1', 'x2', 'y2'])
df.to_csv('/content/drive/MyDrive/trainee/coordinates.csv', header=False,
          index=False, mode='a')
```

```python
import numpy as np
from IPython.display import clear_output
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/trainee/coordinates.csv')
crdt = df.to_numpy(int)
hstgrm, _, _ = np.histogram2d(crdt[:, 2], crdt[:, 3], bins=160)
img = np.uint8((1 - hstgrm / hstgrm.max()) * 255)
img = np.expand_dims(img, axis=-1)
plt.figure(figsize=(16, 16))
plt.imshow(img, cmap='gray')
#plt.xticks([])
#plt.yticks([])
plt.show()
```
